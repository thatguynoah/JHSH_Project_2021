```python
#load in data frame
import pandas as pd

df = pd.read_csv('lung_data.txt', delimiter='\t')
df.head(5)
```

```python
#filtering out NaN
#print("hello")
index = df.index
num_row = len(index)
        
vals = df.loc[pd.isna(df["Gene.symbol"]), :].index

data = {'ID':  [],
        'adj.P.Val': [],
        'P.Value': [],
        't': [],
        'B': [],
        'logFC': [],
        'Gene.symbol': [],
        'Gene.title': []
         }
newer_df = pd.DataFrame(data=data)

for row in range(num_row-47250):
    add = row in vals
    #print(add)
    if (add == False):
        data = {'ID':  [df.iloc[row, 0]],
        'adj.P.Val': [df.iloc[row, 1]],
        'P.Value': [df.iloc[row, 2]],
        't': [df.iloc[row, 3]],
        'B': [df.iloc[row, 4]],
        'logFC': [df.iloc[row, 5]],
        'Gene.symbol': [df.iloc[row, 6]],
        'Gene.title': [df.iloc[row, 7]]
         }
        temp = pd.DataFrame(data=data)
        newer_df = newer_df.append(temp, ignore_index=True)
newer_df.head(5)
```

```python
#selecting top 50
genes_df = newer_df.iloc[0:50, :]

genes_df.head(5)
```

```python
#loading in expression data
exprs_df = pd.read_csv('express_vals.csv')
exprs_df.head(5)
```

```python
print(genes_df.head(3))
print()
print(exprs_df.shape)
```

```python
#reducing expression set to top 50 genes
drops = []
top_exprs_df = exprs_df.copy()
x1, y1 = genes_df.shape
x2, y2 = exprs_df.shape
for rows in range(x1):
    add = genes_df.iloc[rows, 0]
    drops.append(add)

rid = []
for ind in range(x2):
    gate = exprs_df.iloc[ind, 0] in drops#genes_df.iloc[:,0]
    if (gate == False):
        #top_exprs_df = exprs_df.drop([ind])
        rid.append(ind)

top_exprs_df = exprs_df.drop(rid)
```

```python
print(top_exprs_df.shape)
print(drops[0:5])
top_exprs_df.head(3)
```

```python
#get classes
classes_df = pd.read_csv('class.csv')
classes_df.head(20)
```

```python
#create ordered list of classes compared to top_exprs_df
x1, y1 = top_exprs_df.shape
x2, y2 = classes_df.shape
classes = [] #1: MN, 2: BN
temp = list(top_exprs_df.columns)
for col in range(y1-1):
    gsm = temp[col+1]
    for row in range(x2):
        comp = classes_df.iloc[row, 0]
        if (gsm == comp):
            test = classes_df.iloc[row, 1]
            if('MN' in test):
                classes.append(0)
            else:
                classes.append(1)
```

```python
classes[0:10]
```

```python
#creating final df
train_df = top_exprs_df.transpose()

col_names = list(train_df.columns)
x1, y1 = train_df.shape
for col in range(y1):
    change = train_df.iloc[0, col]
    train_df = train_df.rename(columns={col_names[col]:change})
    
train_df = train_df.drop(['Unnamed: 0'])

train_df['class'] = classes
```

```python
train_df.head(3)
```

```python
#import libraries for AI
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create a train test split
x = train_df.drop('class', axis=1)
y = train_df['class']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=5155)
```

```python
#run model
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
#make predictions
y_pred = model.predict(x_test)
print(y_pred)
```

```python
#measure the accuracy
accuracy = accuracy_score(y_test, y_pred)*100
accuracy
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_pred))
```

```python
#best random parameter
best_score = 0
best_parameter = 0
for z in range(10000):
    x = train_df.drop('class', axis=1)
    y = train_df['class']
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=z)
    model = GaussianNB()
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)*100
    if (accuracy > best_score):
        best_score=accuracy
        best_parameter=z
print(best_score, best_parameter)
```
